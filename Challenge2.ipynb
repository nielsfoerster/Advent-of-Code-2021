```python
data=pd.read_csv('data2.txt',names=['name','value'],sep='\s+')
```

```python
data
```

```python
data['forward']=data[data['name']=='forward']['value']
data['down']=data[data['name']=='down']['value']
data['up']=data[data['name']=='up']['value']*(-1)
```

```python
data[data['name']=='down']
```

```python
(sum(data['down'].dropna())+sum(data['up'].dropna()))*sum(data['forward'].dropna())
```

```python
data.fillna(0,inplace=True)
```

```python
data['aim']=data['down']+data['up']
```

```python
data['cum_aim']=data['aim'].cumsum()
```

```python
data['depth']=data['forward']*data['cum_aim']
```
